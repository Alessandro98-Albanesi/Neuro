```python
import cv2
import numpy as np
import pykinect_azure as pykinect



def Blob_detector(im,detector):
    
    
    
    # Detect blobs.
    im_conv = (im/256).astype('uint8')
    #cv2.imshow("Keypoints", im)
    
    #blurred = cv2.medianBlur(im_conv, 5)
    #_filter = cv2.bilateralFilter(blurred, 9, 75, 75)
    #adap_thresh = cv2.adaptiveThreshold(_filter, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV, 21, 0)
    #element = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3)) 
    #dilated = cv2.dilate(adap_thresh, element, iterations=1)
    keypoints = detector.detect(im_conv)
    print(keypoints)
    #cv2.imshow("Keypoints", im_conv)
    
    
    # Draw detected blobs as red circles.
    # cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS ensures the size of the circle corresponds to the size of blob
    im_with_keypoints = cv2.drawKeypoints(im_conv, keypoints, np.array([]), (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

    # Show keypoints
    cv2.imshow("Keypoints", im_with_keypoints)
    



if __name__ == "__main__":

    # Initialize the library, if the library is not found, add the library path as argument
    pykinect.initialize_libraries()

    # Modify camera configuration
    device_config = pykinect.default_configuration
    #device_config.color_format = pykinect.K4A_COLOR_FORMAT_COLOR_BGRA32
    device_config.color_resolution = pykinect.K4A_COLOR_RESOLUTION_720P
    device_config.depth_mode = pykinect.K4A_DEPTH_MODE_NFOV_2X2BINNED
    print(device_config)

    # Start device
    device = pykinect.start_device(config=device_config)
    
    cv2.namedWindow('Infrared Image',cv2.WINDOW_NORMAL)
    
    # Set up the SimpleBlobDetector with default parameters
    params = cv2.SimpleBlobDetector_Params()

    # Set the threshold
    params.minThreshold = 100
    params.maxThreshold = 300

    # Set the area filter
    params.filterByArea = True
    params.minArea = 10
    params.maxArea = 1000

    # Set the circularity filter
    params.filterByCircularity = False
    params.minCircularity = 0.2
    params.maxCircularity = 5

    # Set the convexity filter
    params.filterByConvexity = False
    params.minConvexity = 0.87
    params.maxConvexity = 1

    # Set the inertia filter
    params.filterByInertia = True
    params.minInertiaRatio = 0.01
    params.maxInertiaRatio = 1
    
    # Set the color filter
    params.filterByColor = True
    params.blobColor = 255
    

    # Create a detector with the parameters
    detector = cv2.SimpleBlobDetector_create(params)
    
    
    while True:
   
        # Get capture
        capture = device.update()

        # Get the infrared image
        ret, ir_image = capture.get_ir_image()

        if not ret:
            continue

        Blob_detector(ir_image,detector)
        # Plot imageq
        cv2.imshow('Infrared Image',ir_image)

        # Press q key to stop
        if cv2.waitKey(1) == ord('q'):  
            break
```

```python
# Standard imports
import cv2
import numpy as np;
 
# Read image
im = cv2.imread("ir_image.jpg", cv2.IMREAD_GRAYSCALE)
cv2.imshow("IM", im)
# Set up the detector with default parameters.
detector = cv2.SimpleBlobDetector_create()
 
# Detect blobs.
keypoints = detector.detect(im)
 
# Draw detected blobs as red circles.
# cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS ensures the size of the circle corresponds to the size of blob
im_with_keypoints = cv2.drawKeypoints(im, keypoints, np.array([]), (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
 
# Show keypoints
cv2.imshow("Keypoints", im_with_keypoints)
cv2.waitKey(0)
```
